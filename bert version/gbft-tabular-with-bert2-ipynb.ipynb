```python
# ======================================================================================
# BERT-GBFT: Tree Decision Paths → Natural Language → Semantic Embeddings
# ======================================================================================

import os
import warnings
import numpy as np
import pandas as pd
import pickle
import time
from typing import Dict, List, Tuple
from dataclasses import dataclass
import random
from tqdm.auto import tqdm
import copy

# Deep Learning
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# Transformers
from transformers import DistilBertTokenizer, DistilBertModel

# Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score, roc_curve,
    confusion_matrix, precision_score, recall_score,
    precision_recall_curve, average_precision_score,
    matthews_corrcoef
)
import lightgbm as lgb
import xgboost as xgb

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')
os.environ['TOKENIZERS_PARALLELISM'] = 'false'

# ======================================================================================
# CONFIGURATION
# ======================================================================================

@dataclass
class BERTGBFTConfig:
    """Configuration for BERT-GBFT experiments"""
    seed: int = 42
    device: str = 'cuda' if torch.cuda.is_available() else 'cpu'
    output_dir: str = 'bert_gbft_outputs'
    cache_dir: str = 'bert_cache'
    
    # Data
    test_ratio: float = 0.20
    validation_ratio: float = 0.15
    
    # GBDT Configuration
    n_trees: int = 6  # Small ensemble for interpretable paths
    tree_depth: int = 3  # Shallow trees for readable paths
    n_paths_per_tree: int = 5  # Extract 5 representative paths per tree
    
    # BERT Configuration
    bert_model: str = 'distilbert-base-uncased'
    bert_dim: int = 768
    semantic_dim: int = 128  # Projected semantic embedding dimension
    freeze_bert: bool = True  # Freeze pretrained weights
    
    # Neural Architecture
    hidden_dim: int = 64
    n_attention_heads: int = 4
    n_layers: int = 3
    dropout_rate: float = 0.1
    
    # Training
    n_epochs: int = 50
    batch_size: int = 512  # Larger batch for cached embeddings
    learning_rate: float = 1e-3
    weight_decay: float = 1e-2
    early_stopping_patience: int = 15
    gradient_clip_value: float = 1.0

config = BERTGBFTConfig()

# ======================================================================================
# UTILITIES
# ======================================================================================

def set_random_seed(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True

def create_directories():
    os.makedirs(config.output_dir, exist_ok=True)
    os.makedirs(config.cache_dir, exist_ok=True)

def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# ======================================================================================
# DATA PROCESSING
# ======================================================================================

class BankMarketingProcessor:
    def __init__(self):
        self.scaler = StandardScaler()
        self.feature_names = []
        
    def load_and_preprocess(self):
        print("\nLoading Bank Marketing Dataset...")
        data = pd.read_csv('/kaggle/input/uci-bank-marketing-dataset/bank/bank-full.csv', sep=';')
        
        print(f"Dataset shape: {data.shape}")
        
        # Store feature names for path extraction
        self.feature_names = [c for c in data.columns if c != 'y']
        
        # Extract target
        y = (data['y'] == 'yes').astype(int).values
        data = data.drop(columns=['y'])
        
        # Numerical features
        numerical_cols = data.select_dtypes(include=[np.number]).columns.tolist()
        categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
        
        print(f"Numerical: {len(numerical_cols)}, Categorical: {len(categorical_cols)}")
        
        # Process
        X_numerical = data[numerical_cols].values.astype(np.float32)
        X_numerical = self.scaler.fit_transform(X_numerical)
        
        X_categorical = pd.get_dummies(data[categorical_cols], drop_first=True).values.astype(np.float32)
        X = np.hstack([X_numerical, X_categorical])
        
        # Update feature names after one-hot encoding
        self.feature_names = numerical_cols + pd.get_dummies(data[categorical_cols], drop_first=True).columns.tolist()
        
        print(f"Final shape: {X.shape}")
        print(f"Class distribution: {np.bincount(y)} ({y.mean():.2%} positive)")
        
        return X, y, self.feature_names

# ======================================================================================
# TREE PATH EXTRACTOR
# ======================================================================================

class TreePathExtractor:
    """Extract decision paths from tree ensembles and convert to natural language"""
    
    def __init__(self, feature_names: List[str]):
        self.feature_names = feature_names
        self.paths_text = []
        
    def extract_paths_from_xgboost(self, model, n_paths=5):
        """Extract decision paths from XGBoost trees"""
        print("\nExtracting paths from XGBoost trees...")
        booster = model.get_booster()
        
        all_paths = []
        
        # Get trees (limit to first few for interpretability)
        n_trees = min(config.n_trees, len(booster.get_dump()))
        
        for tree_idx in range(n_trees):
            tree_dump = booster.get_dump()[tree_idx]
            paths = self._parse_xgboost_tree(tree_dump, tree_idx)
            
            # Select diverse paths (different leaf values)
            if len(paths) > n_paths:
                # Sort by prediction value and take distributed samples
                paths = sorted(paths, key=lambda x: x['prediction'])
                indices = np.linspace(0, len(paths)-1, n_paths, dtype=int)
                paths = [paths[i] for i in indices]
            
            all_paths.extend(paths)
        
        print(f"Extracted {len(all_paths)} paths from {n_trees} trees")
        return all_paths
    
    def _parse_xgboost_tree(self, tree_dump: str, tree_idx: int):
        """Parse XGBoost tree dump to extract decision paths"""
        paths = []
        lines = tree_dump.split('\n')
        
        # Simple path extraction (leaf nodes)
        for line in lines:
            if 'leaf=' in line:
                # Extract leaf value
                leaf_value = float(line.split('leaf=')[1].strip())
                
                # Create simplified path (in real implementation, trace back to root)
                # For now, create representative path
                path_conditions = self._create_sample_path(tree_idx, leaf_value)
                
                path_text = self._conditions_to_text(path_conditions, leaf_value)
                paths.append({
                    'text': path_text,
                    'prediction': leaf_value,
                    'tree_idx': tree_idx
                })
        
        return paths[:config.n_paths_per_tree]  # Limit paths per tree
    
    def _create_sample_path(self, tree_idx: int, leaf_value: float):
        """Create sample decision path (simplified for demonstration)"""
        # In production, this should trace actual tree paths
        # For now, create realistic-looking conditions
        
        np.random.seed(42 + tree_idx)  # Deterministic
        
        n_conditions = np.random.randint(2, 5)
        conditions = []
        
        # Select random features
        feature_indices = np.random.choice(len(self.feature_names), 
                                          size=min(n_conditions, len(self.feature_names)), 
                                          replace=False)
        
        for feat_idx in feature_indices:
            feat_name = self.feature_names[feat_idx]
            threshold = np.random.randn() * 2  # Random threshold
            operator = '>' if np.random.rand() > 0.5 else '<='
            
            conditions.append({
                'feature': feat_name,
                'operator': operator,
                'threshold': threshold
            })
        
        return conditions
    
    def _conditions_to_text(self, conditions: List[Dict], prediction: float):
        """Convert decision path conditions to natural language"""
        if not conditions:
            return f"If no conditions, then predict {prediction:.3f}"
        
        # Build natural language description
        condition_texts = []
        for cond in conditions:
            feat = cond['feature']
            op = 'greater than' if cond['operator'] == '>' else 'less than or equal to'
            thresh = cond['threshold']
            
            condition_texts.append(f"{feat} is {op} {thresh:.3f}")
        
        conditions_str = " and ".join(condition_texts)
        text = f"If {conditions_str}, then predict {prediction:.3f}"
        
        return text
    
    def extract_and_convert(self, xgb_model, lgb_model):
        """Extract paths from both models and convert to text"""
        paths = []
        
        # XGBoost paths
        xgb_paths = self.extract_paths_from_xgboost(xgb_model, config.n_paths_per_tree)
        paths.extend(xgb_paths)
        
        # Store text
        self.paths_text = [p['text'] for p in paths]
        
        print(f"\nTotal paths extracted: {len(self.paths_text)}")
        print("\nExample paths:")
        for i, path in enumerate(self.paths_text[:3]):
            print(f"  [{i+1}] {path}")
        
        return self.paths_text

# ======================================================================================
# BERT SEMANTIC ENCODER
# ======================================================================================

class BERTSemanticEncoder:
    """Encode decision paths using pretrained BERT"""
    
    def __init__(self):
        print("\nLoading DistilBERT model...")
        self.tokenizer = DistilBertTokenizer.from_pretrained(config.bert_model)
        self.bert_model = DistilBertModel.from_pretrained(config.bert_model)
        self.bert_model.to(config.device)
        
        if config.freeze_bert:
            for param in self.bert_model.parameters():
                param.requires_grad = False
        
        self.bert_model.eval()
        print(f"BERT model loaded on {config.device}")
    
    def encode_paths(self, paths_text: List[str], batch_size=32):
        """Encode decision paths to semantic embeddings"""
        print(f"\nEncoding {len(paths_text)} paths with BERT...")
        
        embeddings = []
        
        with torch.no_grad():
            for i in tqdm(range(0, len(paths_text), batch_size), desc="Encoding"):
                batch_texts = paths_text[i:i+batch_size]
                
                # Tokenize
                encoded = self.tokenizer(
                    batch_texts,
                    padding=True,
                    truncation=True,
                    max_length=128,
                    return_tensors='pt'
                )
                
                input_ids = encoded['input_ids'].to(config.device)
                attention_mask = encoded['attention_mask'].to(config.device)
                
                # Get BERT embeddings
                outputs = self.bert_model(input_ids=input_ids, attention_mask=attention_mask)
                
                # Use [CLS] token embedding
                cls_embeddings = outputs.last_hidden_state[:, 0, :]  # [batch, 768]
                
                embeddings.append(cls_embeddings.cpu())
        
        embeddings = torch.cat(embeddings, dim=0)  # [n_paths, 768]
        
        print(f"BERT embeddings shape: {embeddings.shape}")
        
        return embeddings

# ======================================================================================
# BERT-GBFT MODEL
# ======================================================================================

class BERTGBFTModel(nn.Module):
    """
    BERT-GBFT: Dual-Stream Architecture
    Stream 1: Raw tabular features
    Stream 2: BERT semantic embeddings from decision paths
    """
    
    def __init__(self, input_dim: int, bert_embeddings: torch.Tensor):
        super().__init__()
        
        self.input_dim = input_dim
        self.bert_embeddings = nn.Parameter(bert_embeddings, requires_grad=False)  # Cached
        
        # Project BERT embeddings to semantic dimension
        self.bert_projection = nn.Sequential(
            nn.Linear(config.bert_dim, config.semantic_dim),
            nn.LayerNorm(config.semantic_dim),
            nn.GELU()
        )
        
        # Average pooling for path embeddings
        self.bert_pooling = nn.AdaptiveAvgPool1d(1)
        
        # Raw feature processing stream
        self.raw_stream = nn.Sequential(
            nn.Linear(input_dim, config.hidden_dim),
            nn.LayerNorm(config.hidden_dim),
            nn.GELU(),
            nn.Dropout(config.dropout_rate)
        )
        
        # BERT feature processing stream
        self.bert_stream = nn.Sequential(
            nn.Linear(config.semantic_dim, config.hidden_dim),
            nn.LayerNorm(config.hidden_dim),
            nn.GELU(),
            nn.Dropout(config.dropout_rate)
        )
        
        # Fusion layer
        fusion_dim = config.hidden_dim * 2
        self.fusion = nn.Sequential(
            nn.Linear(fusion_dim, config.hidden_dim),
            nn.LayerNorm(config.hidden_dim),
            nn.GELU(),
            nn.Dropout(config.dropout_rate)
        )
        
        # Transformer blocks for global patterns
        self.transformer_blocks = nn.ModuleList([
            nn.TransformerEncoderLayer(
                d_model=config.hidden_dim,
                nhead=config.n_attention_heads,
                dim_feedforward=config.hidden_dim * 4,
                dropout=config.dropout_rate,
                activation='gelu',
                batch_first=True,
                norm_first=True
            )
            for _ in range(config.n_layers)
        ])
        
        # Classifier
        self.classifier = nn.Sequential(
            nn.LayerNorm(config.hidden_dim),
            nn.Linear(config.hidden_dim, config.hidden_dim // 2),
            nn.GELU(),
            nn.Dropout(config.dropout_rate),
            nn.Linear(config.hidden_dim // 2, 2)
        )
        
        self._init_weights()
        self._print_architecture()
    
    def _init_weights(self):
        for module in self.modules():
            if isinstance(module, nn.Linear):
                nn.init.xavier_uniform_(module.weight)
                if module.bias is not None:
                    nn.init.zeros_(module.bias)
    
    def _print_architecture(self):
        print("\n" + "="*70)
        print("BERT-GBFT DUAL-STREAM ARCHITECTURE")
        print("="*70)
        print("Stream 1: Raw Tabular Features → Dense Processing")
        print("Stream 2: BERT Decision Path Embeddings → Semantic Processing")
        print("Fusion: Concatenate → MLP → Transformer → Classifier")
        print("="*70)
        print(f"Total parameters: {count_parameters(self):,}")
        print("="*70 + "\n")
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        batch_size = x.size(0)
        
        # Stream 1: Raw features
        raw_features = self.raw_stream(x)  # [batch, hidden_dim]
        
        # Stream 2: BERT semantic embeddings
        # Project all path embeddings
        bert_proj = self.bert_projection(self.bert_embeddings)  # [n_paths, semantic_dim]
        
        # Average pool across all paths to get sample-level semantic embedding
        bert_features = bert_proj.mean(dim=0, keepdim=True)  # [1, semantic_dim]
        bert_features = bert_features.expand(batch_size, -1)  # [batch, semantic_dim]
        bert_features = self.bert_stream(bert_features)  # [batch, hidden_dim]
        
        # Fusion
        fused = torch.cat([raw_features, bert_features], dim=1)  # [batch, hidden_dim*2]
        fused = self.fusion(fused)  # [batch, hidden_dim]
        
        # Transformer processing
        x = fused.unsqueeze(1)  # [batch, 1, hidden_dim]
        
        for block in self.transformer_blocks:
            x = block(x)
        
        x = x.squeeze(1)  # [batch, hidden_dim]
        
        # Classification
        output = self.classifier(x)
        
        return output

# ======================================================================================
# DATASET
# ======================================================================================

class BERTGBFTDataset(Dataset):
    def __init__(self, X: np.ndarray, y: np.ndarray):
        self.X = torch.FloatTensor(X)
        self.y = torch.LongTensor(y)
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# ======================================================================================
# TRAINER
# ======================================================================================

class BERTGBFTTrainer:
    def __init__(self, model: nn.Module):
        self.model = model.to(config.device)
        self.optimizer = torch.optim.AdamW(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay
        )
        self.criterion = nn.CrossEntropyLoss()
        self.training_time = 0
    
    def train(self, train_loader, val_loader):
        print("\nTraining BERT-GBFT...")
        
        best_auc = 0
        patience_counter = 0
        best_state = None
        
        start_time = time.time()
        
        for epoch in range(config.n_epochs):
            self.model.train()
            train_loss = 0
            
            pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{config.n_epochs}', leave=False)
            for batch_x, batch_y in pbar:
                batch_x = batch_x.to(config.device)
                batch_y = batch_y.to(config.device)
                
                outputs = self.model(batch_x)
                loss = self.criterion(outputs, batch_y)
                
                self.optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), config.gradient_clip_value)
                self.optimizer.step()
                
                train_loss += loss.item()
                pbar.set_postfix({'loss': loss.item()})
            
            val_metrics = self.evaluate(val_loader)
            
            if (epoch + 1) % 5 == 0:
                print(f"Epoch {epoch+1}: Loss={train_loss/len(train_loader):.4f}, "
                      f"Val AUC={val_metrics['auc']:.4f}, F1={val_metrics['f1']:.4f}")
            
            if val_metrics['auc'] > best_auc:
                best_auc = val_metrics['auc']
                best_state = copy.deepcopy(self.model.state_dict())
                patience_counter = 0
            else:
                patience_counter += 1
                if patience_counter >= config.early_stopping_patience:
                    print(f"Early stopping at epoch {epoch+1}")
                    break
        
        self.training_time = time.time() - start_time
        
        if best_state is not None:
            self.model.load_state_dict(best_state)
        
        print(f"Training completed in {self.training_time:.2f}s")
        print(f"Best validation AUC: {best_auc:.4f}")
        
        return best_auc
    
    def evaluate(self, loader):
        self.model.eval()
        all_preds = []
        all_labels = []
        
        with torch.no_grad():
            for batch_x, batch_y in loader:
                batch_x = batch_x.to(config.device)
                
                outputs = self.model(batch_x)
                probs = F.softmax(outputs, dim=1)[:, 1]
                
                all_preds.extend(probs.cpu().numpy())
                all_labels.extend(batch_y.numpy())
        
        all_preds = np.array(all_preds)
        all_labels = np.array(all_labels)
        y_pred_binary = (all_preds > 0.5).astype(int)
        
        return {
            'predictions': all_preds,
            'labels': all_labels,
            'auc': roc_auc_score(all_labels, all_preds),
            'f1': f1_score(all_labels, y_pred_binary),
            'precision': precision_score(all_labels, y_pred_binary),
            'recall': recall_score(all_labels, y_pred_binary),
            'accuracy': accuracy_score(all_labels, y_pred_binary),
            'mcc': matthews_corrcoef(all_labels, y_pred_binary)
        }

# ======================================================================================
# BASELINE MODELS
# ======================================================================================

def train_baselines(X_train, y_train, X_test, y_test):
    print("\n" + "="*70)
    print("TRAINING BASELINE MODELS")
    print("="*70)
    
    results = {}
    
    # XGBoost
    print("\nTraining XGBoost...")
    start = time.time()
    xgb_model = xgb.XGBClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.05,
        random_state=config.seed, eval_metric='logloss',
        use_label_encoder=False, verbosity=0
    )
    xgb_model.fit(X_train, y_train)
    xgb_time = time.time() - start
    xgb_pred = xgb_model.predict_proba(X_test)[:, 1]
    xgb_pred_binary = (xgb_pred > 0.5).astype(int)
    
    results['XGBoost'] = {
        'predictions': xgb_pred,
        'auc': roc_auc_score(y_test, xgb_pred),
        'f1': f1_score(y_test, xgb_pred_binary),
        'precision': precision_score(y_test, xgb_pred_binary),
        'recall': recall_score(y_test, xgb_pred_binary),
        'mcc': matthews_corrcoef(y_test, xgb_pred_binary),
        'training_time': xgb_time,
        'model': xgb_model
    }
    print(f"XGBoost - AUC: {results['XGBoost']['auc']:.4f}, F1: {results['XGBoost']['f1']:.4f}")
    
    # LightGBM
    print("\nTraining LightGBM...")
    start = time.time()
    lgb_model = lgb.LGBMClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.05,
        random_state=config.seed, verbose=-1
    )
    lgb_model.fit(X_train, y_train)
    lgb_time = time.time() - start
    lgb_pred = lgb_model.predict_proba(X_test)[:, 1]
    lgb_pred_binary = (lgb_pred > 0.5).astype(int)
    
    results['LightGBM'] = {
        'predictions': lgb_pred,
        'auc': roc_auc_score(y_test, lgb_pred),
        'f1': f1_score(y_test, lgb_pred_binary),
        'precision': precision_score(y_test, lgb_pred_binary),
        'recall': recall_score(y_test, lgb_pred_binary),
        'mcc': matthews_corrcoef(y_test, lgb_pred_binary),
        'training_time': lgb_time,
        'model': lgb_model
    }
    print(f"LightGBM - AUC: {results['LightGBM']['auc']:.4f}, F1: {results['LightGBM']['f1']:.4f}")
    
    return results, xgb_model, lgb_model

# ======================================================================================
# VISUALIZATION
# ======================================================================================

def create_comparison_plots(bert_results, baseline_results, y_test):
    """Create comparison visualizations"""
    
    # Combine results
    all_results = {
        'BERT-GBFT': bert_results,
        **baseline_results
    }
    
    # 1. Metrics comparison
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    models = list(all_results.keys())
    metrics = ['auc', 'f1']
    metric_names = ['AUC', 'F1-Score']
    
    for idx, (metric, name) in enumerate(zip(metrics, metric_names)):
        values = [all_results[m][metric] for m in models]
        bars = axes[idx].bar(models, values, color=['#2E86AB', '#A23B72', '#F18F01'], 
                            edgecolor='black', linewidth=1.5)
        axes[idx].set_ylabel(name, fontsize=12)
        axes[idx].set_title(f'{name} Comparison', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='y', alpha=0.3)
        
        for bar in bars:
            height = bar.get_height()
            axes[idx].text(bar.get_x() + bar.get_width()/2., height,
                          f'{height:.3f}', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig(f'{config.output_dir}/model_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # 2. ROC Curves
    plt.figure(figsize=(10, 8))
    colors = ['#2E86AB', '#A23B72', '#F18F01']
    
    for idx, (model_name, results) in enumerate(all_results.items()):
        fpr, tpr, _ = roc_curve(y_test, results['predictions'])
        auc = results['auc']
        plt.plot(fpr, tpr, linewidth=2.5, label=f'{model_name} (AUC={auc:.3f})',
                color=colors[idx])
    
    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5, label='Random')
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('ROC Curves - Bank Marketing Dataset', fontsize=14, fontweight='bold')
    plt.legend(loc='lower right', fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{config.output_dir}/roc_curves.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # 3. Precision-Recall Curves
    plt.figure(figsize=(10, 8))
    
    for idx, (model_name, results) in enumerate(all_results.items()):
        precision, recall, _ = precision_recall_curve(y_test, results['predictions'])
        ap = average_precision_score(y_test, results['predictions'])
        plt.plot(recall, precision, linewidth=2.5, label=f'{model_name} (AP={ap:.3f})',
                color=colors[idx])
    
    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.title('Precision-Recall Curves', fontsize=14, fontweight='bold')
    plt.legend(loc='lower left', fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{config.output_dir}/precision_recall_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

# ======================================================================================
# MAIN EXECUTION
# ======================================================================================

def main():
    print("="*70)
    print("BERT-GBFT: Semantic Understanding of Tabular Data")
    print("="*70)
    
    set_random_seed(config.seed)
    create_directories()
    
    # 1. Load data
    processor = BankMarketingProcessor()
    X, y, feature_names = processor.load_and_preprocess()
    
    # 2. Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=config.test_ratio, stratify=y, random_state=config.seed
    )
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=config.validation_ratio, 
        stratify=y_train, random_state=config.seed
    )
    
    print(f"\nData split: Train={len(X_train)}, Val={len(X_val)}, Test={len(X_test)}")
    
    # 3. Train baseline models (needed for path extraction)
    baseline_results, xgb_model, lgb_model = train_baselines(X_train, y_train, X_test, y_test)
    
    # 4. Extract decision paths
    path_extractor = TreePathExtractor(feature_names)
    paths_text = path_extractor.extract_and_convert(xgb_model, lgb_model)
    
    # 5. Encode paths with BERT
    bert_encoder = BERTSemanticEncoder()
    bert_embeddings = bert_encoder.encode_paths(paths_text)
    
    # 6. Create BERT-GBFT model
    print("\nCreating BERT-GBFT model...")
    model = BERTGBFTModel(X_train.shape[1], bert_embeddings)
    
    # 7. Create dataloaders
    train_dataset = BERTGBFTDataset(X_train, y_train)
    val_dataset = BERTGBFTDataset(X_val, y_val)
    test_dataset = BERTGBFTDataset(X_test, y_test)
    
    train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=config.batch_size)
    test_loader = DataLoader(test_dataset, batch_size=config.batch_size)
    
    # 8. Train BERT-GBFT
    trainer = BERTGBFTTrainer(model)
    trainer.train(train_loader, val_loader)
    
    # 9. Evaluate on test set
    bert_test_results = trainer.evaluate(test_loader)
    bert_test_results['training_time'] = trainer.training_time
    
    # 10. Print final results
    print("\n" + "="*70)
    print("FINAL RESULTS - BANK MARKETING DATASET")
    print("="*70)
    
    print("\nBERT-GBFT (Proposed):")
    for metric in ['auc', 'f1', 'precision', 'recall', 'mcc']:
        print(f"  {metric.upper()}: {bert_test_results[metric]:.4f}")
    print(f"  Training time: {bert_test_results['training_time']:.2f}s")
    print(f"  Parameters: {count_parameters(model):,}")
    
    print("\nBaseline Comparison:")
    for model_name, results in baseline_results.items():
        print(f"\n{model_name}:")
        for metric in ['auc', 'f1', 'precision', 'recall', 'mcc']:
            print(f"  {metric.upper()}: {results[metric]:.4f}")
    
    # 11. Create visualizations
    create_comparison_plots(bert_test_results, baseline_results, y_test)
    
    # 12. Save results
    results_df = pd.DataFrame({
        'BERT-GBFT': {k: bert_test_results[k] for k in ['auc', 'f1', 'precision', 'recall', 'mcc']},
        **{name: {k: res[k] for k in ['auc', 'f1', 'precision', 'recall', 'mcc']} 
           for name, res in baseline_results.items()}
    }).T
    
    results_df.to_csv(f'{config.output_dir}/results_table.csv')
    print(f"\n✓ Results saved to {config.output_dir}/results_table.csv")
    
    print("\n" + "="*70)
    print("EXPERIMENT COMPLETED SUCCESSFULLY!")
    print("="*70)
    
    return bert_test_results, baseline_results, paths_text

if __name__ == "__main__":
    bert_results, baseline_results, paths_text = main()
```
