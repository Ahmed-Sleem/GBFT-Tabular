```python
# ======================================================================================
# Enhanced Gradient-Boosted Feature Transformer (GBFT)
# With Hierarchical Processing & Comprehensive Complexity Analysis
# ======================================================================================

import os
import warnings
import numpy as np
import pandas as pd
import pickle
import time
from datetime import datetime
from typing import Dict, List, Tuple, Optional, Any
from dataclasses import dataclass
import random
from tqdm.auto import tqdm
import copy
from scipy import stats
from scipy.stats import chi2_contingency, ks_2samp
import psutil
import gc

# Fix for PyTorch compilation issues on Kaggle
os.environ['TORCH_COMPILE_DISABLE'] = '1'
os.environ['PYTORCH_JIT'] = '0'

# Deep Learning
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# Disable torch.compile to avoid DummyDynamo issues
try:
    torch._C._set_graph_executor_optimize(False)
except:
    pass

# Machine Learning
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score, roc_curve,
    confusion_matrix, precision_score, recall_score,
    precision_recall_curve, average_precision_score,
    classification_report, matthews_corrcoef, cohen_kappa_score
)
from sklearn.calibration import calibration_curve
import lightgbm as lgb
import xgboost as xgb
import catboost as cb

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

warnings.filterwarnings('ignore')
plt.style.use('default')

# ======================================================================================
# CONFIGURATION
# ======================================================================================

@dataclass
class ExperimentConfig:
    """Configuration for GBFT experiments"""
    # General
    seed: int = 42
    device: str = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # Directories
    output_dir: str = 'outputs'
    cache_dir: str = 'cache'
    
    # Data
    test_ratio: float = 0.20
    validation_ratio: float = 0.15
    
    # GBDT Configuration
    n_trees: int = 200
    tree_depth: int = 6
    learning_rate_gbdt: float = 0.05
    
    # Neural Architecture (Enhanced)
    hidden_dim: int = 64
    n_attention_heads: int = 4
    n_layers: int = 3
    dropout_rate: float = 0.1
    
    # Training
    n_epochs: int = 50
    batch_size: int = 256
    learning_rate: float = 1e-3
    weight_decay: float = 1e-2
    early_stopping_patience: int = 15
    gradient_clip_value: float = 1.0

config = ExperimentConfig()

# ======================================================================================
# CUSTOM ADAMW OPTIMIZER (to avoid DummyDynamo issues)
# ======================================================================================

class CustomAdamW:
    """
    Custom AdamW optimizer implementation to avoid PyTorch compilation issues
    Implements AdamW: Decoupled Weight Decay Regularization
    """
    def __init__(self, parameters, lr=1e-3, betas=(0.9, 0.999), eps=1e-8, weight_decay=0.01):
        self.parameters = list(parameters)
        self.lr = lr
        self.beta1, self.beta2 = betas
        self.eps = eps
        self.weight_decay = weight_decay
        self.t = 0
        
        # Initialize momentum and velocity
        self.m = [torch.zeros_like(p.data) for p in self.parameters]
        self.v = [torch.zeros_like(p.data) for p in self.parameters]
    
    def zero_grad(self):
        """Zero out gradients"""
        for p in self.parameters:
            if p.grad is not None:
                p.grad.zero_()
    
    def step(self):
        """Update parameters"""
        self.t += 1
        
        for i, p in enumerate(self.parameters):
            if p.grad is None:
                continue
            
            grad = p.grad.data
            
            # Update biased first moment estimate
            self.m[i] = self.beta1 * self.m[i] + (1 - self.beta1) * grad
            
            # Update biased second raw moment estimate
            self.v[i] = self.beta2 * self.v[i] + (1 - self.beta2) * grad.pow(2)
            
            # Compute bias-corrected first moment estimate
            m_hat = self.m[i] / (1 - self.beta1 ** self.t)
            
            # Compute bias-corrected second raw moment estimate
            v_hat = self.v[i] / (1 - self.beta2 ** self.t)
            
            # Update parameters with AdamW (decoupled weight decay)
            # θ_t = θ_{t-1} - lr * (m_hat / (sqrt(v_hat) + eps) + weight_decay * θ_{t-1})
            p.data = p.data - self.lr * (m_hat / (v_hat.sqrt() + self.eps) + self.weight_decay * p.data)
    
    def state_dict(self):
        """Return optimizer state"""
        return {
            't': self.t,
            'lr': self.lr,
            'betas': (self.beta1, self.beta2),
            'eps': self.eps,
            'weight_decay': self.weight_decay,
            'm': self.m,
            'v': self.v
        }
    
    def load_state_dict(self, state_dict):
        """Load optimizer state"""
        self.t = state_dict['t']
        self.lr = state_dict['lr']
        self.beta1, self.beta2 = state_dict['betas']
        self.eps = state_dict['eps']
        self.weight_decay = state_dict['weight_decay']
        self.m = state_dict['m']
        self.v = state_dict['v']

# ======================================================================================
# UTILITIES
# ======================================================================================

def set_random_seed(seed: int):
    """Set random seeds for reproducibility"""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

def create_directories():
    """Create necessary directories"""
    os.makedirs(config.output_dir, exist_ok=True)
    os.makedirs(config.cache_dir, exist_ok=True)
    os.makedirs(f"{config.output_dir}/analysis", exist_ok=True)
    os.makedirs(f"{config.output_dir}/complexity", exist_ok=True)

class DataCache:
    """Simple caching mechanism"""
    
    @staticmethod
    def save(data: Any, name: str):
        path = os.path.join(config.cache_dir, f"{name}.pkl")
        with open(path, 'wb') as f:
            pickle.dump(data, f)
    
    @staticmethod
    def load(name: str) -> Optional[Any]:
        path = os.path.join(config.cache_dir, f"{name}.pkl")
        if os.path.exists(path):
            with open(path, 'rb') as f:
                return pickle.load(f)
        return None

def get_model_size(model):
    """Calculate model size in MB"""
    param_size = 0
    buffer_size = 0
    
    for param in model.parameters():
        param_size += param.nelement() * param.element_size()
    
    for buffer in model.buffers():
        buffer_size += buffer.nelement() * buffer.element_size()
    
    size_mb = (param_size + buffer_size) / 1024 / 1024
    return size_mb

def count_parameters(model):
    """Count trainable parameters"""
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# ======================================================================================
# COMPLEXITY ANALYZER
# ======================================================================================

class ComplexityAnalyzer:
    """Comprehensive complexity analysis for models"""
    
    def __init__(self):
        self.metrics = {}
        
    def analyze_model(self, model, model_name: str, sample_input: torch.Tensor, 
                     n_iterations: int = 100):
        """Comprehensive model analysis"""
        print(f"\nAnalyzing {model_name}...")
        
        metrics = {}
        
        # 1. Parameter count
        if isinstance(model, nn.Module):
            total_params = count_parameters(model)
            metrics['parameters'] = total_params
            metrics['size_mb'] = get_model_size(model)
            
            # Layer-wise analysis
            layer_params = {}
            for name, param in model.named_parameters():
                layer_params[name] = param.numel()
            metrics['layer_params'] = layer_params
        else:
            # For sklearn models
            try:
                model_size = len(pickle.dumps(model)) / 1024 / 1024
                metrics['size_mb'] = model_size
                metrics['parameters'] = 'N/A (Tree-based)'
            except:
                metrics['size_mb'] = 'N/A'
                metrics['parameters'] = 'N/A'
        
        # 2. Inference time
        if isinstance(model, nn.Module):
            model.eval()
            device = next(model.parameters()).device
            sample_input = sample_input.to(device)
            
            # Warm-up
            with torch.no_grad():
                for _ in range(10):
                    _ = model(sample_input)
            
            # Measure inference time
            if torch.cuda.is_available():
                torch.cuda.synchronize()
            
            start_time = time.time()
            
            with torch.no_grad():
                for _ in range(n_iterations):
                    _ = model(sample_input)
            
            if torch.cuda.is_available():
                torch.cuda.synchronize()
            
            end_time = time.time()
            
            avg_inference_time = (end_time - start_time) / n_iterations * 1000  # ms
            metrics['inference_time_ms'] = avg_inference_time
            metrics['throughput_samples_per_sec'] = (len(sample_input) * n_iterations) / (end_time - start_time)
        else:
            # For sklearn models
            sample_np = sample_input.cpu().numpy() if isinstance(sample_input, torch.Tensor) else sample_input
            
            start_time = time.time()
            for _ in range(n_iterations):
                _ = model.predict_proba(sample_np)
            end_time = time.time()
            
            avg_inference_time = (end_time - start_time) / n_iterations * 1000
            metrics['inference_time_ms'] = avg_inference_time
            metrics['throughput_samples_per_sec'] = (len(sample_np) * n_iterations) / (end_time - start_time)
        
        # 3. Memory usage
        if isinstance(model, nn.Module):
            if torch.cuda.is_available():
                torch.cuda.reset_peak_memory_stats()
                with torch.no_grad():
                    _ = model(sample_input)
                metrics['peak_memory_mb'] = torch.cuda.max_memory_allocated() / 1024 / 1024
            else:
                metrics['peak_memory_mb'] = 'N/A (CPU)'
        
        self.metrics[model_name] = metrics
        return metrics
    
    def generate_comparison_report(self):
        """Generate comprehensive comparison report"""
        print("\n" + "="*80)
        print("COMPREHENSIVE COMPLEXITY ANALYSIS")
        print("="*80)
        
        # Create comparison dataframe
        comparison_data = []
        for model_name, metrics in self.metrics.items():
            row = {
                'Model': model_name,
                'Parameters': metrics.get('parameters', 'N/A'),
                'Size (MB)': f"{metrics.get('size_mb', 0):.2f}",
                'Inference (ms)': f"{metrics.get('inference_time_ms', 0):.3f}",
                'Throughput (samples/s)': f"{metrics.get('throughput_samples_per_sec', 0):.0f}",
                'Peak Memory (MB)': metrics.get('peak_memory_mb', 'N/A')
            }
            comparison_data.append(row)
        
        df = pd.DataFrame(comparison_data)
        print("\n" + df.to_string(index=False))
        
        # Save to CSV
        df.to_csv(f"{config.output_dir}/complexity/model_comparison.csv", index=False)
        
        return df
    
    def plot_complexity_comparison(self):
        """Visualize complexity metrics"""
        print("\nGenerating complexity visualizations...")
        
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        
        models = list(self.metrics.keys())
        
        # Filter neural models for parameter count
        neural_models = [m for m in models if isinstance(self.metrics[m].get('parameters'), int)]
        
        # 1. Parameter count
        if neural_models:
            params = [self.metrics[m]['parameters'] for m in neural_models]
            axes[0, 0].bar(neural_models, params, color='skyblue', edgecolor='black', linewidth=1.5)
            axes[0, 0].set_ylabel('Number of Parameters')
            axes[0, 0].set_title('Model Parameters Comparison', fontweight='bold', fontsize=12)
            axes[0, 0].tick_params(axis='x', rotation=45)
            for i, v in enumerate(params):
                axes[0, 0].text(i, v, f'{v:,}', ha='center', va='bottom', fontweight='bold')
            axes[0, 0].grid(axis='y', alpha=0.3)
        
        # 2. Model size
        sizes = [float(self.metrics[m]['size_mb']) if self.metrics[m].get('size_mb') != 'N/A' else 0 
                for m in models]
        axes[0, 1].bar(models, sizes, color='lightcoral', edgecolor='black', linewidth=1.5)
        axes[0, 1].set_ylabel('Size (MB)')
        axes[0, 1].set_title('Model Size Comparison', fontweight='bold', fontsize=12)
        axes[0, 1].tick_params(axis='x', rotation=45)
        for i, v in enumerate(sizes):
            if v > 0:
                axes[0, 1].text(i, v, f'{v:.2f}', ha='center', va='bottom', fontweight='bold')
        axes[0, 1].grid(axis='y', alpha=0.3)
        
        # 3. Inference time
        inf_times = [self.metrics[m].get('inference_time_ms', 0) for m in models]
        axes[1, 0].bar(models, inf_times, color='lightgreen', edgecolor='black', linewidth=1.5)
        axes[1, 0].set_ylabel('Inference Time (ms)')
        axes[1, 0].set_title('Inference Time Comparison', fontweight='bold', fontsize=12)
        axes[1, 0].tick_params(axis='x', rotation=45)
        for i, v in enumerate(inf_times):
            axes[1, 0].text(i, v, f'{v:.3f}', ha='center', va='bottom', fontweight='bold')
        axes[1, 0].grid(axis='y', alpha=0.3)
        
        # 4. Throughput
        throughputs = [self.metrics[m].get('throughput_samples_per_sec', 0) for m in models]
        axes[1, 1].bar(models, throughputs, color='plum', edgecolor='black', linewidth=1.5)
        axes[1, 1].set_ylabel('Throughput (samples/sec)')
        axes[1, 1].set_title('Throughput Comparison', fontweight='bold', fontsize=12)
        axes[1, 1].tick_params(axis='x', rotation=45)
        for i, v in enumerate(throughputs):
            axes[1, 1].text(i, v, f'{v:.0f}', ha='center', va='bottom', fontweight='bold', fontsize=9)
        axes[1, 1].grid(axis='y', alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/complexity/complexity_comparison.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
        
        # Additional: Efficiency plot
        self._plot_efficiency_frontier()
    
    def _plot_efficiency_frontier(self):
        """Plot performance vs complexity trade-off"""
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))
        
        models = list(self.metrics.keys())
        
        # Plot 1: Parameters vs Inference Time
        neural_models = [m for m in models if isinstance(self.metrics[m].get('parameters'), int)]
        if neural_models:
            params = [self.metrics[m]['parameters'] for m in neural_models]
            inf_times = [self.metrics[m].get('inference_time_ms', 0) for m in neural_models]
            
            axes[0].scatter(params, inf_times, s=200, alpha=0.6, edgecolors='black', linewidth=2)
            for i, model in enumerate(neural_models):
                axes[0].annotate(model, (params[i], inf_times[i]), 
                               fontsize=10, ha='center', fontweight='bold')
            axes[0].set_xlabel('Number of Parameters', fontsize=12)
            axes[0].set_ylabel('Inference Time (ms)', fontsize=12)
            axes[0].set_title('Model Complexity: Parameters vs Speed', fontweight='bold', fontsize=12)
            axes[0].grid(True, alpha=0.3)
        
        # Plot 2: Size vs Throughput
        sizes = [float(self.metrics[m]['size_mb']) if self.metrics[m].get('size_mb') != 'N/A' else 0 
                for m in models]
        throughputs = [self.metrics[m].get('throughput_samples_per_sec', 0) for m in models]
        
        axes[1].scatter(sizes, throughputs, s=200, alpha=0.6, c=range(len(models)), 
                       cmap='viridis', edgecolors='black', linewidth=2)
        for i, model in enumerate(models):
            if sizes[i] > 0:
                axes[1].annotate(model, (sizes[i], throughputs[i]), 
                               fontsize=10, ha='center', fontweight='bold')
        axes[1].set_xlabel('Model Size (MB)', fontsize=12)
        axes[1].set_ylabel('Throughput (samples/sec)', fontsize=12)
        axes[1].set_title('Efficiency: Size vs Throughput', fontweight='bold', fontsize=12)
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/complexity/efficiency_frontier.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()

# ======================================================================================
# DATA ANALYSIS
# ======================================================================================

class DatasetAnalyzer:
    """Comprehensive dataset analysis"""
    
    def __init__(self, data: pd.DataFrame, target_col: str = 'income'):
        self.data = data
        self.target_col = target_col
        self.numerical_cols = data.select_dtypes(include=[np.number]).columns.tolist()
        if target_col in self.numerical_cols:
            self.numerical_cols.remove(target_col)
        self.categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
        if target_col in self.categorical_cols:
            self.categorical_cols.remove(target_col)
    
    def generate_full_report(self):
        """Generate comprehensive analysis report"""
        print("\n" + "="*70)
        print("COMPREHENSIVE DATASET ANALYSIS")
        print("="*70)
        
        self._basic_statistics()
        self._target_analysis()
        self._feature_distributions()
        self._correlation_analysis()
        self._missing_values_analysis()
        self._categorical_analysis()
        
        return self
    
    def _basic_statistics(self):
        print("\n1. BASIC STATISTICS")
        print("-" * 70)
        print(f"Total samples: {len(self.data):,}")
        print(f"Total features: {len(self.data.columns)}")
        print(f"  - Numerical: {len(self.numerical_cols)}")
        print(f"  - Categorical: {len(self.categorical_cols)}")
        print(f"\nMemory usage: {self.data.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
    def _target_analysis(self):
        print("\n2. TARGET VARIABLE ANALYSIS")
        print("-" * 70)
        
        if self.target_col in self.data.columns:
            target_dist = self.data[self.target_col].value_counts()
            print(f"\nClass distribution:")
            for cls, count in target_dist.items():
                print(f"  {cls}: {count:,} ({count/len(self.data)*100:.2f}%)")
            
            if len(target_dist) == 2:
                imbalance_ratio = target_dist.max() / target_dist.min()
                print(f"\nImbalance ratio: {imbalance_ratio:.2f}:1")
    
    def _feature_distributions(self):
        print("\n3. FEATURE DISTRIBUTIONS")
        print("-" * 70)
        
        if len(self.numerical_cols) > 0:
            print("\nNumerical features summary:")
            print(self.data[self.numerical_cols].describe().T.to_string())
    
    def _correlation_analysis(self):
        print("\n4. CORRELATION ANALYSIS")
        print("-" * 70)
        
        if len(self.numerical_cols) > 5:
            corr_matrix = self.data[self.numerical_cols].corr()
            high_corr = []
            for i in range(len(corr_matrix.columns)):
                for j in range(i+1, len(corr_matrix.columns)):
                    if abs(corr_matrix.iloc[i, j]) > 0.7:
                        high_corr.append((
                            corr_matrix.columns[i],
                            corr_matrix.columns[j],
                            corr_matrix.iloc[i, j]
                        ))
            
            if high_corr:
                print("\nHighly correlated features (|r| > 0.7):")
                for feat1, feat2, corr in high_corr[:10]:
                    print(f"  {feat1} <-> {feat2}: {corr:.3f}")
            else:
                print("\nNo highly correlated features found (|r| > 0.7)")
    
    def _missing_values_analysis(self):
        print("\n5. MISSING VALUES ANALYSIS")
        print("-" * 70)
        
        missing = self.data.isnull().sum()
        if missing.sum() > 0:
            missing_pct = (missing / len(self.data)) * 100
            missing_df = pd.DataFrame({
                'Missing': missing[missing > 0],
                'Percentage': missing_pct[missing > 0]
            }).sort_values('Missing', ascending=False)
            print(f"\nFeatures with missing values:\n{missing_df.to_string()}")
        else:
            print("\nNo missing values detected")
    
    def _categorical_analysis(self):
        print("\n6. CATEGORICAL FEATURES ANALYSIS")
        print("-" * 70)
        
        if len(self.categorical_cols) > 0:
            print("\nCardinality of categorical features:")
            for col in self.categorical_cols[:10]:
                n_unique = self.data[col].nunique()
                print(f"  {col}: {n_unique} unique values")
    
    def visualize_dataset(self):
        """Create comprehensive visualizations"""
        print("\nGenerating dataset visualizations...")
        
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        
        if self.target_col in self.data.columns:
            target_counts = self.data[self.target_col].value_counts()
            axes[0, 0].bar(target_counts.index.astype(str), target_counts.values, 
                          color=['#3498db', '#e74c3c'], edgecolor='black', linewidth=1.5)
            axes[0, 0].set_title('Target Distribution', fontsize=12, fontweight='bold')
            axes[0, 0].set_xlabel('Class')
            axes[0, 0].set_ylabel('Count')
            for i, v in enumerate(target_counts.values):
                axes[0, 0].text(i, v, f'{v:,}', ha='center', va='bottom', fontweight='bold')
        
        if len(self.numerical_cols) >= 4:
            for idx, col in enumerate(self.numerical_cols[:3]):
                row = (idx + 1) // 2
                col_idx = (idx + 1) % 2
                axes[row, col_idx].hist(self.data[col].dropna(), bins=30, 
                                       color='#2ecc71', edgecolor='black', alpha=0.7)
                axes[row, col_idx].set_title(f'{col} Distribution', fontsize=10, fontweight='bold')
                axes[row, col_idx].set_xlabel(col)
                axes[row, col_idx].set_ylabel('Frequency')
        
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/analysis/dataset_overview.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        if len(self.numerical_cols) > 2:
            plt.figure(figsize=(12, 10))
            corr_matrix = self.data[self.numerical_cols].corr()
            mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
            sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', 
                       cmap='coolwarm', center=0, square=True, 
                       linewidths=1, cbar_kws={"shrink": 0.8})
            plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold', pad=20)
            plt.tight_layout()
            plt.savefig(f"{config.output_dir}/analysis/correlation_matrix.png", dpi=300, bbox_inches='tight')
            plt.show()

# ======================================================================================
# DATA PROCESSING
# ======================================================================================

class TabularDataProcessor:
    """Process tabular data with feature engineering"""
    
    def __init__(self):
        self.numerical_scaler = StandardScaler()
        self.raw_data = None
        
    def load_adult_dataset(self) -> Tuple[np.ndarray, np.ndarray, np.ndarray, pd.DataFrame]:
        """Load and preprocess the Adult dataset"""
        print("\nLoading Adult Income Dataset...")
        
        try:
            data = pd.read_csv('/kaggle/input/adult-census-income/adult.csv')
        except:
            try:
                url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
                columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
                          'marital-status', 'occupation', 'relationship', 'race', 'sex',
                          'capital-gain', 'capital-loss', 'hours-per-week', 
                          'native-country', 'income']
                data = pd.read_csv(url, names=columns, sep=r'\s*,\s*', 
                                 na_values='?', engine='python')
            except:
                print("Could not load dataset")
                return None
        
        self.raw_data = data.copy()
        data = data.dropna().reset_index(drop=True)
        
        y = (data['income'].str.contains('>50K')).astype(int).values
        data = self._engineer_features(data)
        
        numerical_cols = data.select_dtypes(include=[np.number]).columns.tolist()
        if 'income' in numerical_cols:
            numerical_cols.remove('income')
        categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
        if 'income' in categorical_cols:
            categorical_cols.remove('income')
        
        X_numerical = data[numerical_cols].values.astype(np.float32)
        X_numerical = self.numerical_scaler.fit_transform(X_numerical)
        
        X_categorical = pd.get_dummies(data[categorical_cols], drop_first=True).values.astype(np.float32)
        X = np.hstack([X_numerical, X_categorical])
        
        print(f"Dataset shape: {X.shape}")
        print(f"Target distribution: {np.bincount(y)} (Positive ratio: {y.mean():.2%})")
        
        return X, X_numerical, y, self.raw_data
    
    def _engineer_features(self, df: pd.DataFrame) -> pd.DataFrame:
        """Create engineered features"""
        df['age_squared'] = df['age'] ** 2
        df['is_senior'] = (df['age'] > 60).astype(int)
        df['has_capital'] = ((df['capital-gain'] > 0) | (df['capital-loss'] > 0)).astype(int)
        df['net_capital'] = df['capital-gain'] - df['capital-loss']
        df['is_overworked'] = (df['hours-per-week'] > 50).astype(int)
        df['high_education'] = df['education-num'] > 12
        return df

# ======================================================================================
# GBDT FEATURE EXTRACTOR
# ======================================================================================

class GradientBoostedFeatures:
    """Extract features using gradient boosting"""
    
    def __init__(self):
        self.models = []
        
    def fit(self, X: np.ndarray, y: np.ndarray):
        """Train GBDT models"""
        cached = DataCache.load('gbdt_models')
        if cached is not None:
            print("Loading cached GBDT models...")
            self.models = cached
            return self
        
        print("\nTraining Gradient Boosting Ensemble...")
        
        lgb_model = lgb.LGBMClassifier(
            n_estimators=config.n_trees, max_depth=config.tree_depth,
            learning_rate=config.learning_rate_gbdt, subsample=0.8,
            colsample_bytree=0.8, random_state=config.seed, verbose=-1
        )
        lgb_model.fit(X, y)
        self.models.append(lgb_model)
        
        xgb_model = xgb.XGBClassifier(
            n_estimators=config.n_trees, max_depth=config.tree_depth,
            learning_rate=config.learning_rate_gbdt, subsample=0.8,
            colsample_bytree=0.8, random_state=config.seed,
            eval_metric='logloss', use_label_encoder=False, verbosity=0
        )
        xgb_model.fit(X, y)
        self.models.append(xgb_model)
        
        DataCache.save(self.models, 'gbdt_models')
        return self
    
    def transform(self, X: np.ndarray) -> np.ndarray:
        """Extract GBDT features"""
        features = []
        for model in self.models:
            proba = model.predict_proba(X)[:, 1:2]
            features.append(proba)
            if hasattr(model, 'apply'):
                leaves = model.apply(X)
                if len(leaves.shape) > 1:
                    leaves = leaves[:, :10].astype(np.float32) / 100.0
                    features.append(leaves)
        return np.hstack(features).astype(np.float32)

# ======================================================================================
# ENHANCED NEURAL NETWORK COMPONENTS
# ======================================================================================

class LocalPatternExtractor(nn.Module):
    """Extract local patterns from raw features"""
    def __init__(self, input_dim: int, hidden_dim: int, dropout: float = 0.1):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, hidden_dim * 2),
            nn.LayerNorm(hidden_dim * 2),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim * 2, hidden_dim),
            nn.LayerNorm(hidden_dim)
        )
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.layers(x)

class GlobalPatternExtractor(nn.Module):
    """Extract global patterns using attention"""
    def __init__(self, dim: int, n_heads: int = 4, dropout: float = 0.1):
        super().__init__()
        self.n_heads = n_heads
        self.head_dim = dim // n_heads
        
        self.qkv = nn.Linear(dim, dim * 3, bias=False)
        self.projection = nn.Linear(dim, dim)
        self.dropout = nn.Dropout(dropout)
        self.norm = nn.LayerNorm(dim)
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        B, L, D = x.shape
        
        # Multi-head attention
        qkv = self.qkv(x).reshape(B, L, 3, self.n_heads, self.head_dim)
        qkv = qkv.permute(2, 0, 3, 1, 4)
        q, k, v = qkv[0], qkv[1], qkv[2]
        
        scores = torch.matmul(q, k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn = F.softmax(scores, dim=-1)
        attn = self.dropout(attn)
        
        out = torch.matmul(attn, v)
        out = out.permute(0, 2, 1, 3).reshape(B, L, D)
        out = self.projection(out)
        
        return self.norm(out + x)

class HierarchicalFeatureBlock(nn.Module):
    """Hierarchical feature processing with residual connections"""
    def __init__(self, dim: int, n_heads: int = 4, dropout: float = 0.1):
        super().__init__()
        
        self.global_extractor = GlobalPatternExtractor(dim, n_heads, dropout)
        
        self.ffn = nn.Sequential(
            nn.Linear(dim, dim * 4),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(dim * 4, dim)
        )
        self.norm = nn.LayerNorm(dim)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.global_extractor(x)
        x = x + self.dropout(self.ffn(self.norm(x)))
        return x

class EnhancedGradientBoostedFeatureTransformer(nn.Module):
    """
    Enhanced GBFT with Hierarchical Feature Processing:
    Raw Features → Local Patterns → Global Patterns → Decision
    """
    def __init__(self, input_dim: int, gbdt_dim: int):
        super().__init__()
        
        total_dim = input_dim + gbdt_dim
        hidden = config.hidden_dim
        
        # Stage 1: Raw Features → Local Patterns (Dense layers)
        self.local_pattern_extractor = LocalPatternExtractor(
            total_dim, hidden, config.dropout_rate
        )
        
        # Stage 2: Local Patterns → Global Patterns (Attention)
        self.hierarchical_blocks = nn.ModuleList([
            HierarchicalFeatureBlock(hidden, config.n_attention_heads, config.dropout_rate)
            for _ in range(config.n_layers)
        ])
        
        # Stage 3: Feature refinement with residual
        self.feature_refiner = nn.Sequential(
            nn.LayerNorm(hidden),
            nn.Linear(hidden, hidden),
            nn.GELU(),
            nn.Dropout(config.dropout_rate)
        )
        
        # Stage 4: Decision (Classifier)
        self.classifier = nn.Sequential(
            nn.LayerNorm(hidden),
            nn.Linear(hidden, hidden // 2),
            nn.GELU(),
            nn.Dropout(config.dropout_rate),
            nn.Linear(hidden // 2, 2)
        )
        
        self._init_weights()
        self._print_architecture()
    
    def _init_weights(self):
        """Initialize weights with Xavier uniform"""
        for module in self.modules():
            if isinstance(module, nn.Linear):
                nn.init.xavier_uniform_(module.weight)
                if module.bias is not None:
                    nn.init.zeros_(module.bias)
    
    def _print_architecture(self):
        """Print hierarchical architecture"""
        print("\n" + "="*70)
        print("HIERARCHICAL FEATURE PROCESSING ARCHITECTURE")
        print("="*70)
        print("Stage 1: Raw Features → Local Patterns (Dense Layers)")
        print("         └─ Linear → LayerNorm → GELU → Dropout → Linear")
        print("\nStage 2: Local Patterns → Global Patterns (Multi-Head Attention)")
        print(f"         └─ {config.n_layers}x [Attention → FFN → Residual]")
        print("\nStage 3: Feature Refinement (Residual Connection)")
        print("         └─ LayerNorm → Linear → GELU → Dropout")
        print("\nStage 4: Decision (Classifier)")
        print("         └─ LayerNorm → Linear → GELU → Linear → Output")
        print("="*70 + "\n")
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Stage 1: Extract local patterns
        local_features = self.local_pattern_extractor(x)
        
        # Stage 2: Extract global patterns with hierarchical processing
        x = local_features.unsqueeze(1)
        
        for block in self.hierarchical_blocks:
            x = block(x)
        
        x = x.squeeze(1)
        
        # Stage 3: Refine features with residual
        refined_features = self.feature_refiner(x) + local_features
        
        # Stage 4: Classification
        output = self.classifier(refined_features)
        
        return output

# ======================================================================================
# FT-TRANSFORMER
# ======================================================================================

class FTTransformer(nn.Module):
    """FT-Transformer: Feature Tokenizer + Transformer"""
    def __init__(self, input_dim: int, d_token: int = 32, n_blocks: int = 2, 
                 attention_heads: int = 2, ffn_factor: int = 2, dropout: float = 0.1):
        super().__init__()
        self.input_dim = input_dim
        self.d_token = d_token
        
        self.feature_tokenizer = nn.Linear(1, d_token)
        self.cls_token = nn.Parameter(torch.randn(1, 1, d_token))
        self.feature_embeddings = nn.Parameter(torch.randn(input_dim, d_token))
        
        self.blocks = nn.ModuleList([
            nn.TransformerEncoderLayer(
                d_model=d_token,
                nhead=attention_heads,
                dim_feedforward=d_token * ffn_factor,
                dropout=dropout,
                activation='gelu',
                batch_first=True,
                norm_first=True
            )
            for _ in range(n_blocks)
        ])
        
        self.head = nn.Sequential(
            nn.LayerNorm(d_token),
            nn.Linear(d_token, 2)
        )
        
        self._init_weights()
    
    def _init_weights(self):
        for module in self.modules():
            if isinstance(module, nn.Linear):
                nn.init.xavier_uniform_(module.weight)
                if module.bias is not None:
                    nn.init.zeros_(module.bias)
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        batch_size = x.shape[0]
        
        x = x.unsqueeze(-1)
        tokens = self.feature_tokenizer(x)
        tokens = tokens + self.feature_embeddings.unsqueeze(0)
        
        cls_tokens = self.cls_token.expand(batch_size, -1, -1)
        tokens = torch.cat([cls_tokens, tokens], dim=1)
        
        for block in self.blocks:
            tokens = block(tokens)
        
        cls_output = tokens[:, 0]
        return self.head(cls_output)

# ======================================================================================
# TABNET
# ======================================================================================

class AttentiveTransformer(nn.Module):
    """Attentive Transformer for feature selection"""
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.fc = nn.Linear(input_dim, output_dim)
        self.bn = nn.BatchNorm1d(output_dim)
        
    def forward(self, x, prior_scales=None):
        x = self.fc(x)
        x = self.bn(x)
        if prior_scales is not None:
            x = x * prior_scales
        return torch.softmax(x, dim=-1)

class SimpleTabNet(nn.Module):
    """Simplified TabNet for tabular data"""
    def __init__(self, input_dim: int, output_dim: int = 32, n_steps: int = 3):
        super().__init__()
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.n_steps = n_steps
        
        self.initial_bn = nn.BatchNorm1d(input_dim)
        
        self.shared_layers = nn.Sequential(
            nn.Linear(input_dim, output_dim * 2),
            nn.BatchNorm1d(output_dim * 2),
            nn.ReLU()
        )
        
        self.step_layers = nn.ModuleList([
            nn.Sequential(
                nn.Linear(output_dim * 2, output_dim),
                nn.BatchNorm1d(output_dim),
                nn.ReLU()
            )
            for _ in range(n_steps)
        ])
        
        self.attention_layers = nn.ModuleList([
            AttentiveTransformer(output_dim * 2, input_dim)
            for _ in range(n_steps)
        ])
        
        self.head = nn.Sequential(
            nn.Linear(output_dim * n_steps, output_dim),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(output_dim, 2)
        )
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x_normalized = self.initial_bn(x)
        prior_scales = torch.ones(x.size(0), self.input_dim).to(x.device)
        step_outputs = []
        
        for step in range(self.n_steps):
            shared_out = self.shared_layers(x_normalized)
            attention_mask = self.attention_layers[step](shared_out, prior_scales)
            masked_input = x_normalized * attention_mask
            prior_scales = prior_scales * (1 - attention_mask)
            step_out = self.step_layers[step](self.shared_layers(masked_input))
            step_outputs.append(step_out)
        
        aggregated = torch.cat(step_outputs, dim=1)
        return self.head(aggregated)

# ======================================================================================
# DATASET
# ======================================================================================

class TabularDataset(Dataset):
    """Dataset for tabular data"""
    def __init__(self, X_tab: np.ndarray, X_gbdt: np.ndarray, y: np.ndarray):
        self.X = torch.FloatTensor(np.hstack([X_tab, X_gbdt]))
        self.X_tab = torch.FloatTensor(X_tab)
        self.y = torch.LongTensor(y)
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.X_tab[idx], self.y[idx]

# ======================================================================================
# TRAINING
# ======================================================================================

class CosineScheduler:
    """Cosine annealing learning rate scheduler"""
    def __init__(self, optimizer, T_0=10, T_mult=2, eta_min=1e-6):
        self.optimizer = optimizer
        self.T_0 = T_0
        self.T_mult = T_mult
        self.eta_min = eta_min
        self.base_lr = optimizer.lr
        self.T_cur = 0
        self.T_i = T_0
        
    def step(self, epoch_progress):
        """Update learning rate"""
        self.T_cur = epoch_progress
        
        if self.T_cur >= self.T_i:
            self.T_cur = self.T_cur - self.T_i
            self.T_i = self.T_i * self.T_mult
        
        lr = self.eta_min + (self.base_lr - self.eta_min) * (1 + np.cos(np.pi * self.T_cur / self.T_i)) / 2
        self.optimizer.lr = lr

class ModelTrainer:
    """Training pipeline with time tracking"""
    def __init__(self, model: nn.Module, model_name: str = "Model", device: str = None):
        if device is None:
            device = config.device
        self.device = device
        self.model = model.to(self.device)
        self.model_name = model_name
        self.training_time = 0
        
    def train(self, train_loader: DataLoader, val_loader: DataLoader, 
              use_combined_features: bool = True) -> Dict:
        """Train the model"""
        print(f"\nTraining {self.model_name}...")
        
        # Use Custom AdamW optimizer
        optimizer = CustomAdamW(
            self.model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            betas=(0.9, 0.999),
            eps=1e-8
        )
        
        # Cosine annealing scheduler
        scheduler = CosineScheduler(optimizer, T_0=10, T_mult=2, eta_min=1e-6)
        
        criterion = nn.CrossEntropyLoss()
        history = {'train_loss': [], 'val_loss': [], 'val_auc': [], 'val_acc': []}
        best_auc = 0
        patience_counter = 0
        best_state = None
        
        start_time = time.time()
        
        for epoch in range(config.n_epochs):
            self.model.train()
            train_loss = 0
            train_batches = 0
            
            pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{config.n_epochs}', leave=False)
            for batch_idx, batch_data in enumerate(pbar):
                if use_combined_features:
                    batch_x, _, batch_y = batch_data
                else:
                    _, batch_x, batch_y = batch_data
                
                batch_x = batch_x.to(self.device)
                batch_y = batch_y.to(self.device)
                
                outputs = self.model(batch_x)
                loss = criterion(outputs, batch_y)
                
                optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), config.gradient_clip_value)
                optimizer.step()
                
                # Update learning rate
                epoch_progress = epoch + batch_idx / len(train_loader)
                scheduler.step(epoch_progress)
                
                train_loss += loss.item()
                train_batches += 1
                pbar.set_postfix({'loss': loss.item(), 'lr': optimizer.lr})
            
            val_metrics = self.evaluate(val_loader, criterion, use_combined_features)
            
            avg_train_loss = train_loss / train_batches
            history['train_loss'].append(avg_train_loss)
            history['val_loss'].append(val_metrics['loss'])
            history['val_auc'].append(val_metrics['auc'])
            history['val_acc'].append(val_metrics['accuracy'])
            
            if (epoch + 1) % 5 == 0:
                print(f"Epoch {epoch+1}: Train Loss: {avg_train_loss:.4f}, "
                      f"Val AUC: {val_metrics['auc']:.4f}, Val Acc: {val_metrics['accuracy']:.4f}, "
                      f"LR: {optimizer.lr:.6f}")
            
            if val_metrics['auc'] > best_auc:
                best_auc = val_metrics['auc']
                best_state = copy.deepcopy(self.model.state_dict())
                patience_counter = 0
            else:
                patience_counter += 1
                if patience_counter >= config.early_stopping_patience:
                    print(f"Early stopping at epoch {epoch+1}")
                    break
        
        end_time = time.time()
        self.training_time = end_time - start_time
        
        if best_state is not None:
            self.model.load_state_dict(best_state)
        
        print(f"Best validation AUC: {best_auc:.4f}")
        print(f"Training time: {self.training_time:.2f} seconds")
        
        return history
    
    def evaluate(self, loader: DataLoader, criterion=None, 
                 use_combined_features: bool = True) -> Dict:
        """Evaluate the model"""
        if criterion is None:
            criterion = nn.CrossEntropyLoss()
            
        self.model.eval()
        total_loss = 0
        all_preds = []
        all_labels = []
        
        with torch.no_grad():
            for batch_data in loader:
                if use_combined_features:
                    batch_x, _, batch_y = batch_data
                else:
                    _, batch_x, batch_y = batch_data
                    
                batch_x = batch_x.to(self.device)
                batch_y = batch_y.to(self.device)
                
                outputs = self.model(batch_x)
                loss = criterion(outputs, batch_y)
                
                total_loss += loss.item()
                probs = F.softmax(outputs, dim=1)[:, 1]
                
                all_preds.extend(probs.cpu().numpy())
                all_labels.extend(batch_y.cpu().numpy())
        
        all_preds = np.array(all_preds)
        all_labels = np.array(all_labels)
        
        return {
            'loss': total_loss / len(loader),
            'auc': roc_auc_score(all_labels, all_preds),
            'accuracy': accuracy_score(all_labels, (all_preds > 0.5).astype(int)),
            'predictions': all_preds,
            'labels': all_labels
        }

# ======================================================================================
# BASELINE MODELS
# ======================================================================================

def train_baseline_models(X_train: np.ndarray, y_train: np.ndarray,
                         X_test: np.ndarray, y_test: np.ndarray) -> Dict:
    """Train baseline models with time tracking"""
    print("\n" + "="*70)
    print("TRAINING BASELINE MODELS")
    print("="*70)
    
    results = {}
    X_train = X_train.astype(np.float32)
    X_test = X_test.astype(np.float32)
    
    # LightGBM
    print("\n  Training LightGBM...")
    start_time = time.time()
    lgb_model = lgb.LGBMClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.05,
        random_state=config.seed, verbose=-1
    )
    lgb_model.fit(X_train, y_train)
    training_time = time.time() - start_time
    lgb_pred = lgb_model.predict_proba(X_test)[:, 1]
    results['LightGBM'] = {
        'AUC': roc_auc_score(y_test, lgb_pred),
        'Accuracy': accuracy_score(y_test, (lgb_pred > 0.5).astype(int)),
        'predictions': lgb_pred,
        'model': lgb_model,
        'training_time': training_time
    }
    print(f"     Training time: {training_time:.2f}s")
    
    # XGBoost
    print("\n  Training XGBoost...")
    start_time = time.time()
    xgb_model = xgb.XGBClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.05,
        random_state=config.seed, eval_metric='logloss',
        use_label_encoder=False, verbosity=0
    )
    xgb_model.fit(X_train, y_train)
    training_time = time.time() - start_time
    xgb_pred = xgb_model.predict_proba(X_test)[:, 1]
    results['XGBoost'] = {
        'AUC': roc_auc_score(y_test, xgb_pred),
        'Accuracy': accuracy_score(y_test, (xgb_pred > 0.5).astype(int)),
        'predictions': xgb_pred,
        'model': xgb_model,
        'training_time': training_time
    }
    print(f"     Training time: {training_time:.2f}s")
    
    # CatBoost
    print("\n  Training CatBoost...")
    start_time = time.time()
    cat_model = cb.CatBoostClassifier(
        iterations=200, depth=6, learning_rate=0.05,
        random_state=config.seed, verbose=0
    )
    cat_model.fit(X_train, y_train)
    training_time = time.time() - start_time
    cat_pred = cat_model.predict_proba(X_test)[:, 1]
    results['CatBoost'] = {
        'AUC': roc_auc_score(y_test, cat_pred),
        'Accuracy': accuracy_score(y_test, (cat_pred > 0.5).astype(int)),
        'predictions': cat_pred,
        'model': cat_model,
        'training_time': training_time
    }
    print(f"     Training time: {training_time:.2f}s")
    
    return results

# ======================================================================================
# ENHANCED ANALYSIS
# ======================================================================================

class EnhancedAnalyzer:
    """Enhanced model analysis"""
    
    @staticmethod
    def compute_comprehensive_metrics(y_true, y_pred, y_pred_proba):
        """Compute comprehensive evaluation metrics"""
        metrics = {
            'Accuracy': accuracy_score(y_true, y_pred),
            'Precision': precision_score(y_true, y_pred),
            'Recall': recall_score(y_true, y_pred),
            'F1-Score': f1_score(y_true, y_pred),
            'AUC-ROC': roc_auc_score(y_true, y_pred_proba),
            'AUC-PR': average_precision_score(y_true, y_pred_proba),
            'MCC': matthews_corrcoef(y_true, y_pred),
            'Cohen Kappa': cohen_kappa_score(y_true, y_pred)
        }
        return metrics
    
    @staticmethod
    def plot_confusion_matrices(all_results, predictions, y_test):
        """Plot confusion matrices for all models"""
        n_models = min(4, len(predictions))
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        axes = axes.ravel()
        
        for idx, (model_name, pred_data) in enumerate(list(predictions.items())[:4]):
            y_pred = (pred_data['y_pred'] > 0.5).astype(int)
            cm = confusion_matrix(y_test, y_pred)
            
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                       cbar=False, square=True)
            axes[idx].set_title(f'{model_name}', fontweight='bold')
            axes[idx].set_xlabel('Predicted')
            axes[idx].set_ylabel('Actual')
        
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/analysis/confusion_matrices.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
    
    @staticmethod
    def plot_precision_recall_curves(predictions):
        """Plot precision-recall curves"""
        plt.figure(figsize=(10, 8))
        
        colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E']
        for idx, (model_name, pred_data) in enumerate(predictions.items()):
            precision, recall, _ = precision_recall_curve(
                pred_data['y_true'], pred_data['y_pred']
            )
            ap = average_precision_score(pred_data['y_true'], pred_data['y_pred'])
            plt.plot(recall, precision, linewidth=2.5, 
                    label=f'{model_name} (AP={ap:.4f})',
                    color=colors[idx % len(colors)])
        
        plt.xlabel('Recall', fontsize=12)
        plt.ylabel('Precision', fontsize=12)
        plt.title('Precision-Recall Curves', fontsize=14, fontweight='bold')
        plt.legend(loc='lower left', fontsize=10)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/analysis/precision_recall_curves.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
    
    @staticmethod
    def plot_calibration_curves(predictions):
        """Plot calibration curves"""
        n_models = min(4, len(predictions))
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        axes = axes.ravel()
        
        colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E']
        for idx, (model_name, pred_data) in enumerate(list(predictions.items())[:4]):
            fraction_of_positives, mean_predicted_value = calibration_curve(
                pred_data['y_true'], pred_data['y_pred'], n_bins=10
            )
            
            axes[idx].plot([0, 1], [0, 1], 'k--', label='Perfect calibration')
            axes[idx].plot(mean_predicted_value, fraction_of_positives, 
                          marker='o', linewidth=2, color=colors[idx],
                          label=model_name)
            axes[idx].set_xlabel('Mean predicted probability')
            axes[idx].set_ylabel('Fraction of positives')
            axes[idx].set_title(f'{model_name} Calibration', fontweight='bold')
            axes[idx].legend()
            axes[idx].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/analysis/calibration_curves.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()

# ======================================================================================
# VISUALIZATION
# ======================================================================================

def create_visualizations(history_dict: Dict, all_results: Dict, predictions: Dict):
    """Create comprehensive visualizations"""
    print("\nGenerating Visualizations...")
    
    # Training curves comparison
    n_dl_models = len(history_dict)
    if n_dl_models > 0:
        n_rows = (n_dl_models + 1) // 2
        n_cols = min(2, n_dl_models)
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 5*n_rows))
        if n_dl_models == 1:
            axes = [axes]
        else:
            axes = axes.ravel() if n_dl_models > 2 else axes
        
        for idx, (model_name, history) in enumerate(history_dict.items()):
            ax = axes[idx] if n_dl_models > 1 else axes[0]
            epochs = range(1, len(history['train_loss']) + 1)
            ax.plot(epochs, history['val_auc'], 'b-', linewidth=2, marker='o', label='Val AUC')
            ax.plot(epochs, history['val_acc'], 'r-', linewidth=2, marker='s', label='Val Acc')
            ax.set_xlabel('Epoch')
            ax.set_ylabel('Score')
            ax.set_title(f'{model_name} Training Progress', fontweight='bold')
            ax.legend()
            ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f"{config.output_dir}/training_curves_comparison.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
    
    # Model comparison
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    models = list(all_results.keys())
    aucs = [all_results[m]['AUC'] for m in models]
    accs = [all_results[m]['Accuracy'] for m in models]
    
    colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E', '#BC4B51']
    bars1 = axes[0].bar(models, aucs, color=colors[:len(models)], 
                        edgecolor='black', linewidth=1.5)
    for bar in bars1:
        height = bar.get_height()
        axes[0].annotate(f'{height:.4f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3), textcoords="offset points",
                        ha='center', va='bottom', fontsize=10, fontweight='bold')
    axes[0].set_ylabel('AUC Score', fontsize=12)
    axes[0].set_title('Model AUC Comparison', fontsize=14, fontweight='bold')
    axes[0].set_ylim(min(aucs) * 0.95, max(aucs) * 1.05)
    axes[0].grid(True, alpha=0.3, axis='y')
    plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=25, ha='right')
    
    bars2 = axes[1].bar(models, accs, color=colors[:len(models)], 
                        edgecolor='black', linewidth=1.5)
    for bar in bars2:
        height = bar.get_height()
        axes[1].annotate(f'{height:.4f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3), textcoords="offset points",
                        ha='center', va='bottom', fontsize=10, fontweight='bold')
    axes[1].set_ylabel('Accuracy', fontsize=12)
    axes[1].set_title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
    axes[1].set_ylim(min(accs) * 0.95, max(accs) * 1.05)
    axes[1].grid(True, alpha=0.3, axis='y')
    plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=25, ha='right')
    
    plt.tight_layout()
    plt.savefig(f"{config.output_dir}/model_comparison.png", dpi=300, bbox_inches='tight')
    plt.show()
    
    # ROC curves
    plt.figure(figsize=(10, 9))
    colors_roc = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E', '#BC4B51']
    for idx, (model_name, pred_data) in enumerate(predictions.items()):
        fpr, tpr, _ = roc_curve(pred_data['y_true'], pred_data['y_pred'])
        auc = roc_auc_score(pred_data['y_true'], pred_data['y_pred'])
        plt.plot(fpr, tpr, linewidth=2.5, 
                label=f'{model_name} (AUC={auc:.4f})',
                color=colors_roc[idx % len(colors_roc)])
    
    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5, linewidth=1.5, label='Random')
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('ROC Curves Comparison', fontsize=14, fontweight='bold')
    plt.legend(loc='lower right', fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{config.output_dir}/roc_curves.png", dpi=300, bbox_inches='tight')
    plt.show()

# ======================================================================================
# MAIN EXECUTION
# ======================================================================================

def main():
    """Main execution pipeline"""
    print("="*70)
    print(" "*5 + "ENHANCED GRADIENT-BOOSTED FEATURE TRANSFORMER")
    print(" "*10 + "With Hierarchical Processing & Complexity Analysis")
    print("="*70)
    print(f"PyTorch Version: {torch.__version__}")
    print(f"Device: {config.device}")
    print("="*70)
    
    set_random_seed(config.seed)
    create_directories()
    
    # Load and analyze data
    processor = TabularDataProcessor()
    X_full, X_numerical, y, raw_data = processor.load_adult_dataset()
    
    # Dataset analysis
    analyzer = DatasetAnalyzer(raw_data)
    analyzer.generate_full_report()
    analyzer.visualize_dataset()
    
    # Split data
    print("\n" + "="*70)
    print("DATA SPLITTING")
    print("="*70)
    X_train, X_test, y_train, y_test = train_test_split(
        X_full, y, test_size=config.test_ratio,
        stratify=y, random_state=config.seed
    )
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=config.validation_ratio,
        stratify=y_train, random_state=config.seed
    )
    print(f"Train: {len(X_train):,}, Val: {len(X_val):,}, Test: {len(X_test):,}")
    
    # GBDT features
    gbdt_extractor = GradientBoostedFeatures()
    gbdt_extractor.fit(X_train, y_train)
    
    X_train_gbdt = gbdt_extractor.transform(X_train)
    X_val_gbdt = gbdt_extractor.transform(X_val)
    X_test_gbdt = gbdt_extractor.transform(X_test)
    
    # Create datasets
    train_dataset = TabularDataset(X_train, X_train_gbdt, y_train)
    val_dataset = TabularDataset(X_val, X_val_gbdt, y_val)
    test_dataset = TabularDataset(X_test, X_test_gbdt, y_test)
    
    train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_dataset, batch_size=config.batch_size, num_workers=0)
    test_loader = DataLoader(test_dataset, batch_size=config.batch_size, num_workers=0)
    
    input_dim = X_train.shape[1]
    gbdt_dim = X_train_gbdt.shape[1]
    
    print("\n" + "="*70)
    print("TRAINING DEEP LEARNING MODELS")
    print("="*70)
    
    all_results = {}
    predictions = {}
    history_dict = {}
    trained_models = {}
    
    # 1. Enhanced GBFT (Proposed)
    print("\n[1/3] Enhanced GBFT with Hierarchical Processing")
    gbft_model = EnhancedGradientBoostedFeatureTransformer(input_dim, gbdt_dim)
    gbft_trainer = ModelTrainer(gbft_model, "Enhanced GBFT")
    gbft_history = gbft_trainer.train(train_loader, val_loader, use_combined_features=True)
    gbft_metrics = gbft_trainer.evaluate(test_loader, use_combined_features=True)
    
    all_results['Enhanced GBFT'] = {
        'AUC': gbft_metrics['auc'],
        'Accuracy': gbft_metrics['accuracy'],
        'training_time': gbft_trainer.training_time
    }
    predictions['Enhanced GBFT'] = {
        'y_true': y_test,
        'y_pred': gbft_metrics['predictions']
    }
    history_dict['Enhanced GBFT'] = gbft_history
    trained_models['Enhanced GBFT'] = gbft_model
    
    # 2. FT-Transformer
    print("\n[2/3] FT-Transformer")
    ft_model = FTTransformer(input_dim, d_token=32, n_blocks=2, attention_heads=2)
    ft_trainer = ModelTrainer(ft_model, "FT-Transformer")
    ft_history = ft_trainer.train(train_loader, val_loader, use_combined_features=False)
    ft_metrics = ft_trainer.evaluate(test_loader, use_combined_features=False)
    
    all_results['FT-Transformer'] = {
        'AUC': ft_metrics['auc'],
        'Accuracy': ft_metrics['accuracy'],
        'training_time': ft_trainer.training_time
    }
    predictions['FT-Transformer'] = {
        'y_true': y_test,
        'y_pred': ft_metrics['predictions']
    }
    history_dict['FT-Transformer'] = ft_history
    trained_models['FT-Transformer'] = ft_model
    
    # 3. TabNet
    print("\n[3/3] TabNet")
    tabnet_model = SimpleTabNet(input_dim, output_dim=32, n_steps=3)
    tabnet_trainer = ModelTrainer(tabnet_model, "TabNet")
    tabnet_history = tabnet_trainer.train(train_loader, val_loader, use_combined_features=False)
    tabnet_metrics = tabnet_trainer.evaluate(test_loader, use_combined_features=False)
    
    all_results['TabNet'] = {
        'AUC': tabnet_metrics['auc'],
        'Accuracy': tabnet_metrics['accuracy'],
        'training_time': tabnet_trainer.training_time
    }
    predictions['TabNet'] = {
        'y_true': y_test,
        'y_pred': tabnet_metrics['predictions']
    }
    history_dict['TabNet'] = tabnet_history
    trained_models['TabNet'] = tabnet_model
    
    # Train baseline models
    baseline_results = train_baseline_models(X_train, y_train, X_test, y_test)
    
    all_results.update({k: {'AUC': v['AUC'], 'Accuracy': v['Accuracy'], 
                            'training_time': v['training_time']} 
                       for k, v in baseline_results.items()})
    for name, res in baseline_results.items():
        predictions[name] = {'y_true': y_test, 'y_pred': res['predictions']}
        trained_models[name] = res['model']
    
    # ======================================================================================
    # COMPLEXITY ANALYSIS
    # ======================================================================================
    
    print("\n" + "="*70)
    print("COMPLEXITY & PERFORMANCE ANALYSIS")
    print("="*70)
    
    complexity_analyzer = ComplexityAnalyzer()
    
    # Create sample input for analysis
    sample_combined = next(iter(test_loader))[0][:config.batch_size]
    sample_raw = next(iter(test_loader))[1][:config.batch_size]
    
    # Analyze all models
    for model_name, model in trained_models.items():
        if model_name in ['Enhanced GBFT']:
            complexity_analyzer.analyze_model(model, model_name, sample_combined)
        elif model_name in ['FT-Transformer', 'TabNet']:
            complexity_analyzer.analyze_model(model, model_name, sample_raw)
        else:
            # Tree-based models
            complexity_analyzer.analyze_model(model, model_name, X_test[:config.batch_size])
    
    # Generate comparison report
    comparison_df = complexity_analyzer.generate_comparison_report()
    
    # Plot complexity visualizations
    complexity_analyzer.plot_complexity_comparison()
    
    # ======================================================================================
    # ENHANCED ANALYSIS
    # ======================================================================================
    
    print("\n" + "="*70)
    print("ENHANCED ANALYSIS")
    print("="*70)
    
    analyzer = EnhancedAnalyzer()
    analyzer.plot_confusion_matrices(all_results, predictions, y_test)
    analyzer.plot_precision_recall_curves(predictions)
    analyzer.plot_calibration_curves(predictions)
    
    # Comprehensive metrics
    print("\nComprehensive Metrics:")
    detailed_metrics = []
    for model_name, pred_data in predictions.items():
        print(f"\n{model_name}:")
        y_pred = (pred_data['y_pred'] > 0.5).astype(int)
        metrics = analyzer.compute_comprehensive_metrics(
            pred_data['y_true'], y_pred, pred_data['y_pred']
        )
        for metric_name, value in metrics.items():
            print(f"  {metric_name}: {value:.4f}")
        
        metrics['Model'] = model_name
        metrics['Training Time (s)'] = all_results[model_name].get('training_time', 'N/A')
        detailed_metrics.append(metrics)
    
    # Create visualizations
    create_visualizations(history_dict, all_results, predictions)
    
    # ======================================================================================
    # FINAL SUMMARY
    # ======================================================================================
    
    print("\n" + "="*70)
    print("FINAL RESULTS SUMMARY")
    print("="*70)
    
    results_df = pd.DataFrame(all_results).T.sort_values('AUC', ascending=False)
    print("\n" + results_df.to_string())
    
    # Performance analysis
    print("\n" + "="*70)
    print("PERFORMANCE ANALYSIS")
    print("="*70)
    
    gbft_auc = all_results['Enhanced GBFT']['AUC']
    best_baseline_auc = max([v['AUC'] for k, v in all_results.items() 
                             if 'Enhanced GBFT' not in k])
    best_baseline_name = [k for k, v in all_results.items() 
                          if v['AUC'] == best_baseline_auc and 'Enhanced GBFT' not in k][0]
    
    improvement = ((gbft_auc - best_baseline_auc) / best_baseline_auc) * 100
    
    print(f"\nEnhanced GBFT AUC: {gbft_auc:.4f}")
    print(f"Best Baseline ({best_baseline_name}): {best_baseline_auc:.4f}")
    print(f"Improvement: {improvement:+.2f}%")
    
    # Training time analysis
    print("\n" + "="*70)
    print("TRAINING TIME COMPARISON")
    print("="*70)
    for model_name in all_results.keys():
        training_time = all_results[model_name].get('training_time', 'N/A')
        if training_time != 'N/A':
            print(f"{model_name}: {training_time:.2f}s")
    
    # Save results
    results_df.to_csv(f"{config.output_dir}/results.csv")
    
    detailed_df = pd.DataFrame(detailed_metrics)
    detailed_df.to_csv(f"{config.output_dir}/detailed_metrics.csv", index=False)
    
    # Save complexity analysis
    comparison_df.to_csv(f"{config.output_dir}/complexity/complexity_metrics.csv", index=False)
    
    print(f"\n✓ All results saved to {config.output_dir}/")
    print("✓ Experiment completed successfully!")
    
    return all_results, predictions, complexity_analyzer

if __name__ == "__main__":
    try:
        results, predictions, complexity_analyzer = main()
    except Exception as e:
        print(f"\n✗ Error occurred: {str(e)}")
        import traceback
        traceback.print_exc()
```
